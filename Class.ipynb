```python
#Pickle Module

import pickle
'''pickle.dumps() ( short for dump string) creates a binary string representing an object '''

t1 = [1,"two" , 3.0]
s = pickle.dumps(t1)
print(s)
print()

'''The output is a raw binary string that encodes the list t1. Each sumbol encodes one byte. More detail later in the course '''

t2 = pickle.loads(s)
print(t1 == t2)
'''
- pickle.loads(s) takes the binary string s and recreates the original Python object (unpickles it).
- t2 is now a new list with the same contents as t1.
- t1 == t2 → True because both lists have the same elements.
'''
print()
print(t1 is t2)

'''
- is checks whether both variables point to the same object in memory.
- Even though t1 and t2 look identical, they are different objects created at different times.
- So t1 is t2 → False.
'''
print("")
```

```python
import numpy as np

def is_stable(A):
    for i in range(len(A)):
        for j in range(len(A[i])):
            neighbors = 0

            if i + 1 < len(A):
                neighbors += 1
            if j + 1 < len(A[i]):
                neighbors += 1
            if i - 1 >= 0:
                neighbors += 1
            if j - 1 >= 0:
                neighbors += 1

            if A[i][j] <= neighbors:
                return False
    return True
```

```python
A = np.array([
    [5,4,3],
    [4,5,4],
    [3,4,3]
])

is_stable(A)
```
